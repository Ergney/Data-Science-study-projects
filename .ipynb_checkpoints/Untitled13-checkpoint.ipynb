```python
data, target = datasets.make_classification(n_samples=1000, n_features = 5, n_informative = 2, 
                                                      n_classes = 2, n_redundant=0, 
                                                      n_clusters_per_class=1, random_state=5)

data = data.T
```

```python
def sigmoid(z):
    res = 1 / (1 + np.exp(-z))
    return res
```

```python
def calc_logloss(y, y_pred):
    n = 0
    for i in y_pred:
        if i == 0:
            y_pred[n] = 1e-10
        if i == 1:
            y_pred[n] = 1 - 1e-10
        n += 1
    n1 = y * np.log(y_pred)
    n2 = (1.0 - y) * np.log(1.0 - y_pred)
    r = - np.mean(n1 + n2)
    return r
```

```python
def accuracy(y, y_pred):
    n = 0
    for i, j in zip(y, y_pred):
        if i == 1 and j > 0.5:
            n += 1
        elif i == 1 and j <= 0.5:
            n += 1
    result = n/len(y)
    return(result)
```

```python
def eval_model(X, y, iterations, alpha, metric: callable = None):
    np.random.seed(42)
    W = np.random.randn(X.shape[0])
    n = X.shape[1]
    W_dict = {}
    result = pd.DataFrame()
    k = []
    best_err = 9999999
    for j in alpha:
        for i in range(1, iterations+1):
            z = np.dot(W, X)
            y_pred = sigmoid(z)
            err = metric(y, y_pred)
            W -= j * (1/n * np.dot((y_pred - y), X.T))
            #print(i, W, err)
            if err < best_err:
                best_err = err
                best = {"iteration": i, "alpha": j, "error": best_err, "W": W}
            k.append(err)
            
        result[str(j)] = k
        k = []
        W_dict[str(j)] = W
        W = np.random.randn(X.shape[0])
        
    #print(i, W, err)
    #if i % (iterations / 10) == 0:
    
    return result, W_dict, best
```

```python
#W = eval_model(X, y, iterations=500, alpha=6e-7)
alpha_1 = [i/100 for i in range(1,11)]
alpha_2 = [i/1000 for i in range(1,11)]
alpha_3 = [i/10000 for i in range(1,11)]
#alpha_1, alpha_2, alpha_3
r_1, W_1, best_1 = eval_model(data, target, iterations=2000, alpha=alpha_1, metric=calc_logloss)
r_2, W_2, best_2 = eval_model(data, target, iterations=2000, alpha=alpha_2, metric=calc_logloss)
r_3, W_3, best_3 = eval_model(data, target, iterations=2000, alpha=alpha_3, metric=calc_logloss)
```

```python
best_3
```

```python
def graph(data, alpha_list, best):
    
    rcParams["figure.figsize"] = 16, 10
    
    print(f'Best parametrs: \n iteration:{best["iteration"]} \n alpha:{best["alpha"]} \n W:{best["W"]} \n error:{best["error"]}')
    
    for i in alpha_list:
        plt.plot(range(len(data[str(i)])), data[str(i)], label=f'alpha:{i}')
        
        
        
         
    
    plt.legend()
    
    plt.xlabel("Количество итераций")
    plt.ylabel("log_loss")
    
    plt.show()
```

```python
graph(r_1, alpha_1, best_1)
graph(r_2, alpha_2, best_2)
graph(r_3, alpha_3, best_3)
```

4. Создайте функцию calc_pred, возвращающую предсказанный класс (на вход подаются W, который уже посчитан функцией eval_model и X, на выходе - массив y_pred).

```python
def calc_pred(W, data):
    z = np.dot(W, data)
    result = sigmoid(z)
    for i, j in zip(result, range(len(result))):
        if i <= 0.5:
            result[j] = 1
        else:
            result[j] = 0
    return result
```
