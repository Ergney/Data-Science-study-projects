```python
import numpy as np
from sklearn import model_selection
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
X, y = load_iris(return_X_y=True)

# Для наглядности возьмем только первые два признака (всего в датасете их 4)
X = X[:, :2]
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
cmap = ListedColormap(['red', 'green', 'blue'])
plt.figure(figsize=(7, 7))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap)
```

```python
def e_metrics(x1, x2):
    
    distance = 0
    for i in range(len(x1)):
        distance += np.square(x1[i] - x2[i])
      
    return np.sqrt(distance)
```

```python
def knn(x_train, y_train, x_test, k, weighing=None):
    
    answers = []
    for x in x_test:
        test_distances = []
            
        for i in range(len(x_train)):
            
            # расчет расстояния от классифицируемого объекта до
            # объекта обучающей выборки
            distance = e_metrics(x, x_train[i])
            
            # Записываем в список значение расстояния и ответа на объекте обучающей выборки
            test_distances.append((distance, y_train[i]))
        
        # создаем словарь со всеми возможными классами
        classes = {class_item: 0 for class_item in set(y_train)}
        
        # Сортируем список и среди первых k элементов подсчитаем частоту появления разных классов
        t = sorted(test_distances)[0:k]
        coef = [1/2**i for i in range(1,k+1)]
        
        #Если применено взвешивание по порядковому номеру
        if weighing == "number":
            for d, c in zip(t, coef):
                classes[d[1]] += 1 * c
                               
        #Если применено взвешивание по расстоянию
        elif weighing == "distance":
            for d in sorted(test_distances)[0:k]:
                classes[d[1]] += d[0]
                
        #Если применено взвешивание не применялось
        else:
            for d in sorted(test_distances)[0:k]:
                classes[d[1]] += 1
        
        

            
        # Записываем в список ответов наиболее часто встречающийся класс
        answers.append(sorted(classes, key=classes.get)[-1])
    return answers
```

```python
def get_graph(X_train, y_train, k, weighing=None):
    cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA','#00AAFF'])

    h = .02

    # Расчет пределов графика
    x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
    y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    # Получим предсказания для всех точек
    Z = knn(X_train, y_train, np.c_[xx.ravel(), yy.ravel()], k, weighing=weighing)

    # Построим график
    Z = np.array(Z).reshape(xx.shape)
    plt.figure(figsize=(7,7))
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Добавим на график обучающую выборку
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cmap)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title(f"Трехклассовая kNN классификация при k = {k}")
    plt.show()
```

```python
def accuracy(pred, y):
    return (sum(pred == y) / len(y))
```

```python
y_pred1 = knn(X_train, y_train, X_test, 5)
y_pred2 = knn(X_train, y_train, X_test, 5, weighing = "number")
y_pred3 = knn(X_train, y_train, X_test, 5, weighing = "distance")

print(f'Точность алгоритма при k = {5}: {accuracy(y_pred1, y_test):.3f} Без взвешивания')
get_graph(X_train, y_train, k)
print(f'Точность алгоритма при k = {5}: {accuracy(y_pred2, y_test):.3f} С взвешиванием по номеру соседа')
get_graph(X_train, y_train, k, weighing = "number")
print(f'Точность алгоритма при k = {5}: {accuracy(y_pred3, y_test):.3f} С взвешиванием по расстоянию')
get_graph(X_train, y_train, k, weighing = "distance")
```

```python
k_list = [i for i in range(1,16)]

best_standart = {}
best_number = {}
best_distance = {}

for i in k_list:
    y_pred1 = knn(X_train, y_train, X_test, k=i)
    y_pred2 = knn(X_train, y_train, X_test, k=i, weighing = "number")
    y_pred3 = knn(X_train, y_train, X_test, k=i, weighing = "distance")
    
    best_standart[accuracy(y_pred1, y_test)] = i
    best_number[accuracy(y_pred2, y_test)] = i
    best_distance[accuracy(y_pred3, y_test)] = i

print(f'Оптимальное количество соседей для модели без взвешивания: {best_standart[(max(best_standart))]}, Точность: {max(best_standart)}')
print(f'Оптимальное количество соседей для модели с взвешиванием по номеру соседа: {best_number[(max(best_number))]}, Точность: {max(best_number)}')
print(f'Оптимальное количество соседей для модели с взвешиванием по расстоянию: {best_distance[(max(best_distance))]}, Точность: {max(best_distance)}')
```
