–°—É–¥—è –ø–æ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–º—É –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é –≤ –≤—ã–±–æ—Ä–∫–µ –∏–º—é—Ç—Å—è –≤—ã–±—Ä–æ—Å—ã, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–≥—É—Ç –Ω–µ–≥–∞—Ç–∏–≤–Ω–æ —Å–∫–∞–∑–∞—Ç—å—Å—è –Ω–∞ –∞–Ω–∞–ª–∏–∑–µ –¥–∞–Ω–Ω—ã—Ö.

–ò–∑–±–∞–≤–∏–º—Å—è –æ—Ç –Ω–∏—Ö, –æ—Ç–±—Ä–æ—Å–∏–≤ –≤—Å–µ –∑–Ω–∞—á–µ–Ω–∏—è –º–µ–Ω–µ–µ 0.1 –∫–≤–∞–Ω—Ç–∏–ª—è –∏ –±–æ–ª–µ–µ 0.9

```python
def vibrosi(atribut):
    data = df[atribut].loc[(df[atribut] >= df[atribut].quantile(0.05)) & (df[atribut] <= df[atribut].quantile(0.95))] 
    return data
```

```python
df["views"] = vibrosi("views")
df["likes"] = vibrosi("likes")
df["dislikes"] = vibrosi("dislikes")
df["comment_count"] = vibrosi("comment_count")
```

–¢–µ–ø–µ—Ä—å –ø—Ä–æ–≤–µ—Ä–∏–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –≤–µ–ª–∏—á–∏–Ω (views, likes, dislikes, comment_count) –∏–∑ –¥–∞—Ç–∞—Å–µ—Ç–∞.

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
fig.set_size_inches(15, 4)

ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(12, 8)
fig.subplots_adjust(wspace=0.4, hspace=0.4)

ax1.hist(df["views"], bins=15)
ax1.set_title("views")

ax2.hist(df["likes"], bins=15)
ax2.set_title("likes")

ax3.hist(df["dislikes"], bins=15)
ax3.set_title("dislikes")

ax4.hist(df["comment_count"], bins=15)
ax4.set_title("comment_count")

plt.show()
```
