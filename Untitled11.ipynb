```python
import numpy as np
import pandas as pd
from sklearn import datasets
import itertools
import matplotlib.pyplot as plt
#from sklearn.linear_model import LinearRegression, SGDClassifier, SGDRegressor
from pylab import rcParams
```

```python
data, target, coef = datasets.make_regression(n_samples=100, n_features = 4, n_targets = 1,
                                              noise = 5, coef = True, random_state = 2)
```

```python
def calc_mse(y, y_pred):
    err = np.mean((y - y_pred)**2)
    return err
```

```python
def standart_gradient(x, y, iterations=1000, alpha=0.0001, lmbd=1e-2, L1=False, L2=False):
    result_list_mse = []
    result_list_W = []
    count = 0
    n = len(data)
    W = np.array([1.0,1.0,1.0,1.0])
    for i in range(iterations):
        y_pred = np.dot(x, W)
        for ii in range(len(W)):
            count += 1
            coef = 0
            if L1 == True:
                coef = lmbd * sum(abs(W))/2
            elif L2 == True:
                coef = lmbd * sum(W**2)/2
            W -= 2 * alpha * (1/n * (np.dot(x.T, (y_pred - y)) + coef))
            
        error = calc_mse(y, y_pred)
        
        result_list_mse.append(error)
        result_list_W.append(W)
    return result_list_mse, result_list_W, count
```

```python
def eval_sgd_model(data, target, iterations=1000, alpha=0.0001, lmbd=1e-2, L1=False, L2=False):
    result_list_mse = []
    result_list_W = []
    W = np.array([1.0,1.0,1.0,1.0])
    n = len(data)
    count = 0
    for i in range(iterations):
        for j, k in itertools.product(range(n), range(len(W))):
            count += 1
            coef = 0
            y_pred = np.dot(data, W)
            if L1 == True:
                coef = lmbd * sum(abs(W))/2
            elif L2 == True:
                coef = lmbd * sum(W**2)/2
            W[k] -= 2 * alpha * np.dot(data[j].T, (np.dot(data[j], W[k]) - target[j])) + coef / n
            if count % len(W) == 0:

                error = calc_mse(target, y_pred)
                                      
        result_list_mse.append(error)
        result_list_W.append(W)
    return result_list_mse, result_list_W.append, count
        
```

```python
MSE_1, W_1, c_1 = standart_gradient(data, target)
MSE_2, W_2, c_2 = standart_gradient(data, target, L1=True)
MSE_3, W_3, c_3 = standart_gradient(data, target, L2=True)

MSE_4, W_4, c_4 = eval_sgd_model(data, target)
MSE_5, W_5, c_5 = eval_sgd_model(data, target, L1=True)
MSE_6, W_6, c_6 = eval_sgd_model(data, target, L2=True)
```

```python
rcParams["figure.figsize"] = 10, 8

plt.plot(range(1, len(MSE_1)+1), MSE_1, label="   ")
plt.plot(range(1, len(MSE_2)+1), MSE_2, label="  c L1 ")
plt.plot(range(1, len(MSE_3)+1), MSE_3, label="  c L2 ")

plt.plot(range(1, len(MSE_4)+1), MSE_4, label="   ")
plt.plot(range(1, len(MSE_5)+1), MSE_5, label="   L1 ")
plt.plot(range(1, len(MSE_6)+1), MSE_6, label="   L2 ")

plt.xlabel(" ")
plt.ylabel("MSE")

plt.legend()

plt.show()
```
