```python
import numpy as np
import pandas as pd
from sklearn import datasets
import itertools
import matplotlib.pyplot as plt
#from sklearn.linear_model import LinearRegression, SGDClassifier, SGDRegressor
from pylab import rcParams
```

```python
data, target, coef = datasets.make_regression(n_samples=500, n_features = 5, n_targets = 1,
                                              noise = 5, coef = True, random_state = 28)
```

```python
def calc_mse(y, y_pred, W, regul=None, _lambda=1e-2, L1=0, L2=0):
    if regul=="L1":
        L1 = _lambda * sum(abs(W))/2
    elif regul=="L2":
        L2 = _lambda * sum(W**2)/2   
    err = np.mean((y - y_pred)**2) + L1 + L2
    return err
```

```python
#Нормализация

def norm(data):
    n1, n2  = data.shape[0], data.shape[1]
     
    data = data.T
    for i in range(n2):
        min_data = data[i].min()
        max_data = data[i].max()
        for j in range(n1):
            data[i][j] = (data[i][j] - min_data)/(max_data - min_data)
            
    return data.T

#Стандартизация

def standart(data):
    n1, n2  = data.shape[0], data.shape[1]
    
    data = data.T
    for i in range(n2):
        mean = np.mean(data[i])
        stdev = np.sqrt(sum((data[i]-mean)**2)/n1)
        
        for j in range(n1):
            data[i][j] = (data[i][j] - mean)/stdev
    
    return data.T
```

```python
def standart_gradient(x, y, iterations=10000, alpha=0.0001, regul=None):
    result_list_mse = []
    iters = []
    #result_list_W = []
    count = 0
    n = len(data)
    W = np.ones(data[0].shape, dtype=float)
    for i in range(iterations):
        y_pred = np.dot(x, W)
        count += 1
        W -= 2 * alpha * (1/n * (np.dot(x.T, (y_pred - y))))
        
        if regul == "L1":    
            error = calc_mse(y, y_pred, W, regul="L1")
        elif regul == "L2":
            error = calc_mse(y, y_pred, W, regul="L2")
        else:
            error = calc_mse(y, y_pred, W)
        
        
        if count % 20 == 0:
            result_list_mse.append(error)
            iters.append(count)
        
        #result_list_W.append(W)
    return [iters, result_list_mse] #result_list_W, count
```

```python
#Стохастический градинентный спуск 1.0

def eval_sgd_model_2(x, y, iterations=50000, alpha=0.01, regul=None):
    result_list_mse = []
    iters = []
    #result_list_W = []
    count = 0
    n = len(data)
    W = np.ones(data[0].shape, dtype=float)
    for i in range(iterations):
        count += 1
        index = np.random.randint(n)
        y_pred = np.dot(x[index], W)
        W -= 2 * alpha * (1/n * np.dot(x[index].T, (y_pred - y[index])))
        
        if regul == "L1":    
            error = calc_mse(y[index], y_pred, W, regul="L1")
        elif regul == "L2":
            error = calc_mse(y[index], y_pred, W, regul="L2")
        error = calc_mse(y[index], y_pred, W)
        
        if count % 50 == 0:
            result_list_mse.append(error)
            iters.append(count)
            
        #result_list_W.append(W)
    return [iters, result_list_mse] #result_list_W, count
```

```python
#Данные с нормализацией
norm_data = norm(data)
#Данные с стандартизацией
standart_data = standart(data)

#Стандартный градиентный спуск.

#Стоковые данные без регуляризации
result_1 = standart_gradient(data, target)

#Нормализванные данные без регуляризации
result_1n = standart_gradient(norm_data, target)

#Стандартизованные данные без регуляризации
result_1s = standart_gradient(standart_data, target)

#Стоковые данные c L1 регуляризацией
result_1_L1 = standart_gradient(data, target, regul="L1")

#Нормализванные данные c L1 регуляризацией
result_1n_L1 = standart_gradient(data, target, regul="L1")

#Стандартизованные данные c L1 регуляризацией
result_1s_L1 = standart_gradient(data, target, regul="L1")

#Стоковые данные c L2 регуляризацией
result_1_L2 = standart_gradient(data, target, regul="L2")

#Нормализванные данные c L2 регуляризацией
result_1n_L2 = standart_gradient(data, target, regul="L2")

#Стандартизованные данные c L2 регуляризацией
result_1s_L2 = standart_gradient(data, target, regul="L2")


#Стохастический градиентный спуск.

#Стоковые данные без регуляризации
result_2 = eval_sgd_model_2(data, target)

#Нормализванные данные без регуляризации
result_2n = eval_sgd_model_2(norm_data, target)

#Стандартизованные данные без регуляризации
result_2s = eval_sgd_model_2(standart_data, target)

#Стоковые данные c L1 регуляризацией
result_2_L1 = eval_sgd_model_2(data, target, regul="L1")

#Нормализванные данные c L1 регуляризацией
result_2n_L1 = eval_sgd_model_2(data, target, regul="L1")

#Стандартизованные данные c L1 регуляризацией
result_2s_L1 = eval_sgd_model_2(data, target, regul="L1")

#Стоковые данные c L2 регуляризацией
result_2_L2 = eval_sgd_model_2(data, target, regul="L2")

#Нормализванные данные c L2 регуляризацией
result_2n_L2 = eval_sgd_model_2(data, target, regul="L2")

#Стандартизованные данные c L2 регуляризацией
result_2s_L2 = eval_sgd_model_2(data, target, regul="L2")
```
